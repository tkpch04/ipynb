```python
import pandas as pd

# Misalnya kamu membaca file CSV
df = pd.read_csv('heart_attack_vaccine_data.csv')

# Sekarang baru bisa memanggil kolom 'Type of Record'
print(df['Vaccine Dose'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Vaccine Dose', order=df['Vaccine Dose'].value_counts().index)
plt.title('Jumlah Pasien per Jenis Dosis Vaksin')
plt.xlabel('Jumlah')
plt.ylabel('Jenis Dosis')
plt.show()
```
